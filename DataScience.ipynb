<h2 class="h2">What is Data Science?</h2>
<p>Data science is a deep study of the massive amount of data, which involves extracting meaningful insights from raw, structured, and unstructured data that is processed using the scientific method, different technologies, and algorithms.</p>
<p>It is a multidisciplinary field that uses tools and techniques to manipulate the data so that you can find something new and meaningful. </p>
<p>Data science uses the most powerful hardware, programming systems, and most efficient algorithms to solve the data related problems. It is the future of artificial intelligence. </p>
<p>In short, we can say that data science is all about:</p>
<ul class="points">
<li>Asking the correct questions and analyzing the raw data.</li>
<li>Modeling the data using various complex and efficient algorithms.</li>
<li>Visualizing the data to get a better perspective.</li>
<li>Understanding the data to make better decisions and finding the final result. </li>
</ul>

<h2 class="h2">Data Science Lifecycle</h2>
<p>The life-cycle of data science is explained as below diagram. </p>
<img src="https://static.javatpoint.com/tutorial/data-science/images/data-science-lifecycle.png" alt="Data Science tutorial" />
<p>The main phases of data science life cycle are given below:</p>
<p><strong>1. Discovery:</strong> The first phase is discovery, which involves asking the right questions. When you start any data science project, you need to determine what are the basic requirements, priorities, and project budget. In this phase, we need to determine all the requirements of the project such as the number of people, technology, time, data, an end goal, and then we can frame the business problem on first hypothesis level.</p>
<p><strong>2. Data preparation:</strong> Data preparation is also known as Data Munging. In this phase, we need to perform the following tasks:</p>
<ul class="points">
<li>Data cleaning </li>
<li>Data Reduction</li>
<li>Data integration</li>
<li>Data transformation,</li>
</ul>
<p>After performing all the above tasks, we can easily use this data for our further processes.</p>
<p><strong>3. Model Planning:</strong> In this phase, we need to determine the various methods and techniques to establish the relation between input variables. We will apply Exploratory data analytics(EDA) by using various statistical formula and visualization tools to understand the relations between variable and to see what data can inform us. Common tools used for model planning are:</p>
<ul class="points">
<li>SQL Analysis Services</li>
<li>R</li>
<li>SAS</li>
<li>Python</li>
</ul>
<p><strong>4. Model-building:</strong> In this phase, the process of model building starts. We will create datasets for training and testing purpose. We will apply different techniques such as association, classification, and clustering, to build the model. </p>
<p>Following are some common Model building tools:</p>
<ul class="points">
<li>SAS Enterprise Miner</li>
<li>WEKA</li>
<li>SPCS Modeler</li>
<li>MATLAB</li>
</ul>
<p><strong>5. Operationalize:</strong> In this phase, we will deliver the final reports of the project, along with briefings, code, and technical documents. This phase provides you a clear overview of complete project performance and other components on a small scale before the full deployment.</p>
<p><strong>6. Communicate results:</strong> In this phase, we will check if we reach the goal, which we have set on the initial phase. We will communicate the findings and final result with the business team.</p>

<h2>How Does a Data Scientist Work?</h2>
<p>A Data Scientist requires expertise in several 
backgrounds: </p>
<ul>
  <li>Machine Learning</li>
  <li>Statistics</li>
  <li>Programming (Python or R)</li>
  <li>Mathematics</li>
  <li>Databases</li>
</ul>
<p>A Data Scientist must find patterns within the data. Before he/she can find 
the patterns, he/she must organize the data in a standard format.</p>
<p>Here is how a Data Scientist works:</p>
<ol>
  <li><strong>Ask the right questions</strong> - To understand the business 
  problem.</li>
  <li><strong>Explore and collect data</strong> - From database, web logs, 
  customer feedback, etc.</li>
  <li><strong>Extract the data</strong> - Transform the data to a standardized 
format.</li>
  <li><strong>Clean the data</strong> - Remove erroneous values from the data.</li>
  <li><strong>Find and replace missing values</strong> - Check for 
missing values and replace them with a suitable value (e.g. an average value).</li>
  <li><strong>Normalize data</strong> - Scale the values in a practical range 
  (e.g. 140 cm is smaller than 1,8 m. However, the number 140 
is larger than 1,8. - so scaling is important).</li>
  <li><strong>Analyze data, find patterns and make future predictions</strong>.</li>
  <li><strong>Represent the result </strong>- Present the result with useful 
  insights in a way the &quot;company&quot; can understand.</li>
</ol>
